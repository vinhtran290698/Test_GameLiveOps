```python
# Import library module
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')

import warnings
warnings.filterwarnings('ignore')
```

The seriated position is not important, so drop the number of this string.

```python
# Read excel file
df = pd.read_excel(r'D:\DATA ANALYST COURSES\DATA\GameLiveOpTest.xlsx')

# Create a function to drop special key of Position column
def drop_position(x):
    count = 0
    for i in range(0,len(x)):
        if x[i] == '_':
            count = count + 1
        a = i
    if count >= 1:
        return x[:a-1] 
    else: return x[:]

# Use apply task to handle every row
df.Position = df.Position.apply(drop_position)
df
```

```python
salary_position = pd.pivot_table(data=df,index='Position',values='Salary',aggfunc=['count',np.mean,np.sum],fill_value=0,margins=True,margins_name='Total').sort_values(by='Position')
salary_position.style.background_gradient(cmap='coolwarm')
```

```python
salary_position['count', 'Salary'].iloc[:-1].sort_values(ascending=False).plot(kind='bar');
```

```python
salary_position['mean', 'Salary'].iloc[:-1].sort_values(ascending=False).plot(kind='bar');
```

```python
plt.boxplot(salary_position['mean', 'Salary'].iloc[:-1]);
```

```python
salary_position['sum', 'Salary'].iloc[:-1].sort_values(ascending=False).plot(kind='bar')
```

```python
plt.figure(figsize=(10,10))
salary_position['sum', 'Salary'].iloc[:-1].sort_values(ascending=False).plot(kind='pie',autopct="%.2f",fontsize=15)
plt.show()
```
